## Описание данных

Таблица **users** (информация о пользователях):
- user_id — уникальный идентификатор пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя
- age — возраст пользователя (годы)
- reg_date — дата подключения тарифа (день, месяц, год)
- churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- city — город проживания пользователя
- tarif — название тарифного плана

Таблица **calls** (информация о звонках):
- id — уникальный номер звонка
- call_date — дата звонка
- duration — длительность звонка в минутах
- user_id — идентификатор пользователя, сделавшего звонок

Таблица **messages** (информация о сообщениях):
- id — уникальный номер сообщения
- message_date — дата сообщения
- user_id — идентификатор пользователя, отправившего сообщение

Таблица **internet** (информация об интернет-сессиях):
- id — уникальный номер сессии
- mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
- session_date — дата интернет-сессии
- user_id — идентификатор пользователя

Таблица **tariffs** (информация о тарифах):
- tariff_name — название тарифа
- rub_monthly_fee — ежемесячная абонентская плата в рублях
- minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
- messages_included — количество сообщений в месяц, включённых в абонентскую плату
- mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

### Признак "id"

Визуальный анализ первых пяти строк таблицы "df_messages" позволяет предположить, что данный признак сформирован аналогичным образом как в таблице "df_calls".

Проверим данную гипотезу, попробуем найти записи, где первое число признака "id" отличается от уникального идентификатора пользователя "user_id".

```python
df_messages[~df_messages['id'].str.split('_', expand=True).iloc[:, 0].astype(int) == df_messages['user_id']]
```

Таких записей нет, значит во всех записях первое число дублирует признак "id".

Также проведём дополнительную проверку второго числа признака: если второе число является нумерацией сообщений для каждого клиента, то последний номер должен отражать количество сообщений каждого клиента. Аналогия - последний номер квартиры отражает количество квартир в доме. Сумма всех номеров должна равняться размеру таблицы.

```python
(
    # Обращаемся к признаку "id".
    df_messages['id']
    # Обращаемся как к строке.
    .str
    # Выполняем разбиение по символу нижнего подчёркивания.
    .split('_', expand=True)
    # Рассматриваем полученные подстроки как числа.
    .astype(int)
    # Выполняем группировку по первому столбцу - уникальному идентификатору пользователя.
    .groupby(by=0)[1]
    # Выбираемый максимальный номер сообщения для каждого пользователя.
    .max()
    # Увеличиваем на единицу, так как нумерация начинается с нуля.
    .apply(lambda x: x + 1)
    # Суммируем все номера.
    .sum()
)
```

Сумма всех номеров равна длинне таблицы, значит принимаем гипотезу, что второе число - номер сообщения для каждого клиента.

#### Расчёт абонентской платы по месяцам.

Рассчитаем абонентскую плату для каждого пользователя с учётом месяцев подключения и отключения тарифа.

```python
# Заимствуем таблицу для формирования таблицы абоненской платы.
df_tariffs_status = df_internet_month.copy()

# Обнулим все значения.
df_tariffs_status.iloc[:, :] = np.nan

# Для каждого столбца:
for column in df_tariffs_status.columns:
    # ячейки таблицы, когда был подключен тариф, заполним единицами,
    df_tariffs_status.loc[df_users['reg_date'].dt.month <= column, [column]] = 1
    # ячейки таблицы, когда был отключен тариф, заполним нулями.
    df_tariffs_status.loc[df_users['churn_date'].dt.month < column, [column]] = np.nan
```

Выполним рассчёт абоненской платы для каждого пользователя по месяцам.

```python
# За основу возьмём таблицу-карту "df_tariffs", где указаны месяцы, когда был подключен тариф.
df_fee_tariffs = df_tariffs_status.copy()

# Добавление в таблицу статуса тарифа столбца "стоимость тарифа в месяц".
df_fee_tariffs['rub_monthly_fee'] = df_users_fee['rub_monthly_fee']

# Умножение всех статусов на стоимость тарифа:
# 0 - когда тариф не был подключен, останется 0.
# 1 - когда тариф был подключен, станет стоимостью тарифа в месяц.
for column in df_fee_tariffs.columns:
    df_fee_tariffs.loc[:, [column]] = df_fee_tariffs.loc[:, [column]].apply(lambda x: 
                                                                            x * df_fee_tariffs.loc[:, 'rub_monthly_fee'])

# Удаление столбца "стоимость тарифа в месяц".
df_fee_tariffs = df_fee_tariffs.drop(columns=['rub_monthly_fee'])
```

- из каждого месячного значения вычесть лимит тарифа;
- обнулить отрицательные значения;
- положительные значения умножить на тарифную стоимость минуты звонка сверх лимита.

```python
# Возьмём за основу таблицу о звонках "df_calls_month", где указана продолжительность звонков по месяцам.
df_fee_calls = df_calls_month.copy()

# Добавим поле о количестве включённых минут в тариф "minutes_included".
df_fee_calls['minutes_included'] = (
    df_users[['tariff']]
    .merge(right=df_tariffs[['tariff_name', 'minutes_included']],
           how='left',
           left_on ='tariff',
           right_on='tariff_name')[['minutes_included']]
    .set_index(df_users.index)
)

# Добавим поле о стоимости минуты сверх лимита тарифа "rub_per_minute".
df_fee_calls['rub_per_minute'] = (
    df_users[['tariff']]
    .merge(right=df_tariffs[['tariff_name', 'rub_per_minute']],
           how='left',
           left_on ='tariff',
           right_on='tariff_name')[['rub_per_minute']]
    .set_index(df_users.index)
)
```

Таблица "df_fee_calls" содержит информацию о количестве потраченных минут в месяц, количестве минут в тарифе и стоимость минуты сверх тарифного плана для каждого клиента.

Вычислим ежемесячные затраты клиентов на дополнительные минуты.

```python
# Список столбцов "месяцы".
columns_month = df_fee_calls.columns[:-2]

# Вычитание соответствующего лимита тарифа по месяцам для каждого клиента.
for column in columns_month:
    df_fee_calls.loc[:, [column]] = df_fee_calls.loc[:, [column]].apply(lambda x: x - df_fee_calls.loc[:, 'minutes_included'])

# Обнуление отрицательных значений, так как в этом случае лимит не был превышен.
df_fee_calls[columns_month] = df_fee_calls[columns_month].mask(df_fee_calls[columns_month] < 0, 0)

# Умножение минут вне лимита на ставку тарифа.
for column in columns_month:
    df_fee_calls.loc[:, [column]] = df_fee_calls.loc[:, [column]].apply(lambda x: x * df_fee_calls.loc[:, 'rub_per_minute'])
```

Теперь из необходимо вычесть лимиты тарифа и умножить на тарифную ставку, чтобы получить затраты на дополнительные услуги оператора.

```python
# Возьмём за основу таблицу о звонках "df_internet_month", где указан объём интернет-трафика по месяцам.
df_fee_internet = df_internet_month.copy()

# Добавим поле о количестве включённых гигабайт в тариф "gb_per_month_included".
df_fee_internet['gb_per_month_included'] = (
    df_users[['tariff']]
    .merge(right=df_tariffs[['tariff_name', 'gb_per_month_included']],
           how='left',
           left_on ='tariff',
           right_on='tariff_name')[['gb_per_month_included']]
    .set_index(df_users.index)
)

# Добавим поле о стоимости гигабайт сверх лимита тарифа "rub_per_message".
df_fee_internet['rub_per_gb'] = (
    df_users[['tariff']]
    .merge(right=df_tariffs[['tariff_name', 'rub_per_gb']],
           how='left',
           left_on ='tariff',
           right_on='tariff_name')[['rub_per_gb']]
    .set_index(df_users.index)
)

# Список столбцов "месяцы".
columns_month = df_fee_internet.columns[:-2]

# Вычитание соответствующего лимита тарифа по месяцам для каждого клиента.
for column in columns_month:
    df_fee_internet.loc[:, [column]] = (
        df_fee_internet
        .loc[:, [column]]
        .apply(lambda x: x - df_fee_internet.loc[:, 'gb_per_month_included'])
    )
    
# Обнуление отрицательных значений, так как в этом случае лимит не был превышен.
df_fee_internet[columns_month] = df_fee_internet[columns_month].mask(df_fee_internet[columns_month] < 0, 0)

# Умножение гигабайт вне лимита на ставку тарифа.
for column in columns_month:
    df_fee_internet.loc[:, [column]] = df_fee_internet.loc[:, [column]].apply(lambda x:
                                                                              x * df_fee_internet.loc[:, 'rub_per_gb'])

# Удаление лишних столбцов.
df_fee_internet = df_fee_internet.drop(columns=['gb_per_month_included', 'rub_per_gb'])
```

### Проверка гипотезы

**Формулирование нулевой и альтернативной гипотез**

Для проверки гипотезы, что средние выручки пользователей тарифов «ultra» и «smart» различаются, необходимо сформулировать нулевую гипотезу H0, которая будет предполагаться изначально верной, и альтернативную гипотезу H1, которая будет принята в случае отвержения гипотезы H0.

В качестве нулевой гипотезы H0 примем предположение, что средние двух выборок равны:
$$H0: M(smart) = M(ultra)$$

В качестве альтернативной гипотезы H1 примем предположение, что средние двух выборок отличны. Используем двухсторонний тест, так как теоретически средние выборок могут располагаться как угодно относительно друг друга:
$$H1: M(smart) \neq M(ultra)$$

**Выбор критерия**

Тестируемые выборки являются независимыми - клиенты, использующие разные тарифы никак не связаны друг с другом. Также анализ значений выборок показал серьёзные отличия стандартных отклонений и, как следствие, дисперсий.

Резюмируя вышеизложенное, для тестирования гипотезы о равенстве средних платежей необходимо использовать метод <code>.ttest_ind()</code> для независимых выборок с настройкой параметра <code>equal_var=False</code>.

Выполним тестирование.

```python
# result_fee_tariffs = st.stats.ttest_ind(a=df_fee_smart, b=df_fee_ultra, equal_var=False)
result_fee_tariffs = st.ttest_ind(a=df_fee_smart, b=df_fee_ultra, equal_var=False)

p_value = result_fee_tariffs.pvalue[0]
```

```python
if p_value > alpha:
    print(f'Принимаем нулевую гипотезу.')
elif p_value < alpha:
    print(f'Отвергаем нулевую гипотезу.')
else:
    print(f'Необходим иной метод проверки.')

print(f'Уровень статистической значимости p-value:\t{p_value}')
```

Полученное значение p-value крайне мало, значительно меньше установленного порога альфа 5%. Результат тестирования гипотезы о равенстве средних платежей двух выборок свидетельствует о том, что при верности гипотезы H0 шансы встретить равные значения крайне малы, практически равны нулю.

Таким образом, мы отвергаем гипотезу H0 о равенстве средних выборочных и принимаем альтернативную гипотезу H1 о различие средних двух выборок.
