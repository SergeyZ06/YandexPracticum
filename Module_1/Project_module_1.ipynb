### Изучение пропусков признака "Critic_Score"

#### Проверка зависимости пропусков "Critic_Score" от платформы

```python
# Вывод платформ с соответствующими долями заполненных значений признака "Critic_Score" и количеством записей.
(
df_games[['Platform', 'Critic_Score']]
    .groupby(by='Platform')
    .apply(lambda df: df[['Critic_Score']].notnull().mean().apply('{:.2%}'.format))
    .sort_values(by='Critic_Score', ascending=False)
    .join(pd.Series(df_games[['Platform']].value_counts(), name='amount_of_records')
          .reset_index()
          .set_index('Platform'))
)
```

Пропуски присутствуют у всех платформ, при этом у части платформ оценки критиков отсутствуют полностью. Платформы без рейтинга малочисленны в сравнении с платформами с рейтингом.

```python
plt.figure(figsize=(16, 6))

sns.histplot(df_games.loc[df_games['Critic_Score'].isna(), ['Year_of_Release']],
             bins=36,
             label='Игры без оценок критиков',
             alpha=0.4,
             palette=sns.color_palette('Reds', n_colors=1))

sns.histplot(df_games.loc[~df_games['Critic_Score'].isna(), ['Year_of_Release']],
             bins=31,
             label='Игры с оценками критиков',
             alpha=0.6)

plt.xticks(range(int(df_games['Year_of_Release'].min()),
                 int(df_games['Year_of_Release'].max()) + 1,
                 1))

plt.xticks(rotation=45)

plt.title('График распределения года выпуска игр с оценками критиков и без', fontsize=16)
plt.xlabel('Год выпуска', fontsize=14)
plt.ylabel('Количество объектов', fontsize=14)

plt.legend(prop={'size': 16});
```

До 2000 года у игр практически не было оценок критиков.

После 2000 ситуация стала неоднозначной: у игр стали появляться оценки критиков, но в зависимости от года доли игр с оценками и без меняли. Так например с 2001 по 2008 года в релизе преобладали игры с рейтингом, затем ситуация изменилась на противоположную.

### Признак "year_of_release"

```python
# Вывод первых пяти объектов с пропусками в признаке "year_of_release".
df_games[df_games['year_of_release'].isna()].head()
```

```python
# Вывод количества записей с пропусками в признаке "year_of_release".
df_games['year_of_release'].isna().sum()
```

Как было установлено ранее, одни и те же игры могли выходить на разных платформах. Исходя из предположения, что игры выходят на разных платформах примерно в одно время, пропуски года выпуска могут быть заполнены с группировкой по названию игр.

```python
# Сохранение индекса записей с пропусками в признаке "year_of_release".
index = df_games[df_games['year_of_release'].isna()].index

# Заполнение пропусков медианными значениями с группировкой.
df_games['year_of_release'] = df_games['year_of_release'].fillna(
    df_games.groupby(by='name')['year_of_release'].transform('median')
)
```

```python
# Проверка: вывод количества пропусков в признаке "year_of_release".
df_games['year_of_release'].isna().sum()
```

Более чем для 100 записей удалось заполнить пропуски при помощи группировки по именам.

В остальных случаях корректное заполнение невозможно.

Ввиду того, что для решения задачи необходим анализ с разбиением по годам, игры без даты выхода не представляют информационной ценности. Доля таких игр составляет около 1% от всей выборки, что позволяет исключить их из дальнейшего рассмотрения.

```python
# Исключение объектов с пропусками в признаке "name".
df_games = df_games[~df_games['year_of_release'].isna()]
```

```python
# Проверка: вывод количества пропусков в признаке "year_of_release".
df_games['year_of_release'].isna().sum()
```

# Проверка гипотез

Выполним проверки следующих гипотез:
- средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- средние пользовательские рейтинги жанров Action и Sports разные.

Пороговый уровень статистической значимости альфа установим 5%, это одно из общепринятых значений для тестирования гипотез. Данного значения будет достаточно для обнаружения умеренных различий.

```python
alpha = 0.05
```

## Проверка гипотезы о рейтингах платформ Xbox One и PC

### Предварительный анализ

```python
# Таблица пользовательских рейтингов платформы "Xbox One".
df_user_score_xone = (
df_games
    .loc[(df_games['platform'] == 'XOne') & (~df_games['user_score'].isna()), ['user_score']]
    .rename(columns={'user_score': 'user_score_xone'})
)

# Таблица пользовательских рейтингов платформы "PC".
df_user_score_pc = (
df_games
    .loc[(df_games['platform'] == 'PC') & (~df_games['user_score'].isna()), ['user_score']]
    .rename(columns={'user_score': 'user_score_pc'})
)

# Вывод основных статистик двух платформ.
df_user_score_xone.describe().join(
    df_user_score_pc.describe()
)
```

```python
plt.figure(figsize=(16, 5))

sns.histplot(df_user_score_pc, bins=21, label='PC', alpha=0.3)
sns.histplot(df_user_score_xone, bins=20, label='XOne', palette=sns.color_palette('Reds', n_colors=1), alpha=0.6)

plt.legend(prop={'size': 16})
plt.xticks(np.arange(0, 10.5, 0.5))

plt.title(label='График взаимного расположения гистограмм пользовательских оценок платформ "XOne" и "PC"', fontsize=16)
plt.xlabel(xlabel='Оценки пользователей', fontsize=14)
plt.ylabel(ylabel='Количество игр', fontsize=14);
```

Предварительный анализ показывает, что выборки пользовательских оценок для платформ "XOne" и "PC" значительно отличаются по размеру, последняя меньше почти в 4 раза.

Основные статистики достаточно близки:
- средние значения двух выборок отличаются на пол балла рейтинга;
- стандартное отклонение отличается примерно на 10%;
- минимальные, максимальные значения и значения квартилей схожи.

На графике распределения двух выборок видно, что из-за разницы размеров одна выборка полностью покрывает другую. Выборки обладают схожими характерами распределений.

### Проверка гипотезы

**Формулирование нулевой и альтернативной гипотез**

Для проверки гипотезы, что средние пользовательские рейтинги платформ Xbox One и PC одинаковы, необходимо сформулировать нулевую гипотезу H0, которая будет предполагаться изначально верной, и альтернативную гипотезу H1, которая будет принята в случае отвержения гипотезы H0.

В качестве нулевой гипотезы H0 примем предположение, что средние двух выборок равны:
$$H0: M(XOne) = M(PC)$$

В качестве альтернативной гипотезы H1 примем предположение, что средние двух выборок отличны. Используем двухсторонний тест, так как теоретически средние выборок могут располагаться как угодно относительно друг друга:
$$H1: M(XOne) \neq M(PC)$$

**Выбор критерия**

Тестируемые выборки являются независимыми - бескорыстные отзывы клиентов разных платформ не должны быть связаны друг с другом. Анализ значений выборок показал сходства стандартных отклонений и, как следствие, дисперсий.

Резюмируя вышеизложенное, для тестирования гипотезы о равенстве средних оценок необходимо использовать метод <code>.ttest_ind()</code> для независимых выборок с настройкой параметра <code>equal_var=True</code>.

Выполним тестирование.

```python
result_user_score_platforms = st.ttest_ind(a=df_user_score_xone, b=df_user_score_pc, equal_var=True)

p_value = result_user_score_platforms.pvalue[0]
```

```python
if p_value > alpha:
    print(f'Принимаем нулевую гипотезу.')
elif p_value < alpha:
    print(f'Отвергаем нулевую гипотезу.')
else:
    print(f'Необходим иной метод проверки.')

print(f'Уровень статистической значимости p-value:\t{p_value}')
```

Полученное значение p-value крайне мало, значительно меньше установленного порога альфа 5%. Результат тестирования гипотезы о равенстве средних пользовательских оценок двух выборок свидетельствует о том, что при верности гипотезы H0 шансы встретить равные значения крайне малы, практически равны нулю.

Таким образом, мы отвергаем гипотезу H0 о равенстве средних выборочных и принимаем альтернативную гипотезу H1 о различие средних двух выборок.

**Средние пользовательские рейтинги платформ «Xbox One» и «PC» различаются.**

3. **Положительные отзывы.** Как критиков, так и простых игроков. Исследование показало, что игры с большим числом реализованных копий как правило встречаются среди игр с высокими оценками. Бывают исключения на подобии игр без точных оценок, но в целом тенденция прослеживается.


4. **Рейтинг игры.** Данная особенность косвенно связана с объёмами продаж. Чаще всего пользователи выбирают либо игры без рейтинга вообще, либо с рейтингом "E". Данные категории являются самыми массовыми, а значит популярными, следует сделать ставку на них.


5. **Объёмы продаж.** В отдельных случаях были выявлены бестселлеры, объёмы продаж которых значительно превосходят рядовые показатели. Рекламные кампания должна учитывать и эти явления, чтобы магазин не упустил прибыль от самых популярных продуктов.


Анализ накопленной статистики позволил выявить вышеописанные закономерности. Однако тенденции рынка могут меняться, поэтому данные особенности нельзя принимать за универсальный шаблон. Для каждого нового периода времени необходимо проводить дополнительные исследования истории продаж.
