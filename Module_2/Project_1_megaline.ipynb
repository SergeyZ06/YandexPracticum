### Логистическая регрессия

#### Подготовка данных

Линейные модели лучше работают с стандартизованными данными, поэтому проведём соответствующую подготовку данных.

```python
# Инициализация инструмента стандартизации.
ss = StandardScaler()

# Стандартизация выборок.
X_train_ss = ss.fit_transform(X_train)
X_valid_ss = ss.transform(X_valid)
X_test_ss = ss.transform(X_test)
```

Также для лучшей работы линейных моделей необходимо наличие перехвата - единичного признака, который модель самостоятельно оценит как смещение.

#### Полиномизация

```python
# Функция для выполнения полиномизации второй степени.
def polynomization_2(dataframe_original, bias=True, drop_original_features=False):
    dataframe = dataframe_original.copy()
    list_features_1 = dataframe.columns
    list_features_2 = list_features_1.copy()
    
    for feature_1 in list_features_1:
        for feature_2 in list_features_2:
            dataframe[f'{feature_1}_{feature_2}'] = dataframe[feature_1] * dataframe[feature_2]
        if len(list_features_2) > 1:
            list_features_2 = list_features_2[1:]
    
    if bias:
        dataframe['bias'] = 1
    
    if drop_original_features:
        dataframe = dataframe.drop(columns=list_features_1)
    
    return dataframe
```

```python
# Формирование нового пространства признаков обучающей выборки для полиномизации.
X_train_ss_p2 = (
    X_train_ss
    .copy()
    .drop(columns=['intercept'])
)

# Выполнение полиномизации.
X_train_ss_p2 = polynomization_2(X_train_ss_p2)

# Проверка.
X_train_ss_p2.info()
```

```python
# Формирование нового пространства признаков валидационной выборки для полиномизации.
X_valid_ss_p2 = (
    X_valid_ss
    .copy()
    .drop(columns=['intercept'])
)

# Выполнение полиномизации.
X_valid_ss_p2 = polynomization_2(X_valid_ss_p2)

# Проверка.
X_valid_ss_p2.info()
```

```python
# Формирование нового пространства признаков тестовой выборки для полиномизации.
X_test_ss_p2 = (
    X_test_ss
    .copy()
    .drop(columns=['intercept'])
)

# Выполнение полиномизации.
X_test_ss_p2 = polynomization_2(X_test_ss_p2)

# Проверка.
X_test_ss_p2.info()
```

#### Обучение и оценка базовой модели

```python
# Инициализация логистической модели без настроек.
lr = LogisticRegression(random_state=random_state_global)

# Обучение модели.
lr.fit(X_train_ss, y_train['is_ultra'])

# Выполнение прогноза на обучающей выборке.
y_pred_train = lr.predict(X_train_ss)

# Выполнение прогноза на валидационной выборке.
y_pred_valid = lr.predict(X_valid_ss)
```

```python
# Проверка метрик качества на обучающей выборке.
print(classification_report(y_train, y_pred_train))
```

```python
# Проверка метрик качества на валидационной выборке.
print(classification_report(y_valid, y_pred_valid))
```

```python
# Набор гиперпараметров для проверки.
params = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [50, 100, 150],
    'min_samples_leaf': [5, 10, 15],
    'class_weight': [None, 'balanced', {0: 0.3, 1: 0.7}, {0: 0.4, 1: 0.6}],
    'random_state': [random_state_global]
}

# Инициализация инструмента для поиска лучших параметров.
# gs_rfc = GridSearchCV(estimator=RandomForestClassifier(),
#                       param_grid=params,
#                       scoring='accuracy',
#                       n_jobs=-1,
#                       verbose=0)

gs_rfc = RandomizedSearchCV(estimator=RandomForestClassifier(),
                            param_distributions=params,
                            scoring='accuracy',
                            n_jobs=-1,
                            verbose=0,
                            random_state=random_state_global)
```
