### –¢–µ—Ö–Ω–∏–∫–∞ "upsampling"

#### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è upsampling.
def upsampling(X, y):
    df = X.copy()
    df['y'] = y.copy()
    
    # –ï—Å–ª–∏ –Ω—É–ª–µ–≤–æ–π –∫–ª–∞—Å—Å –ø—Ä–µ–æ–±–ª–∞–¥–∞–µ—Ç,
    if len(df[df['y'] == 0]) > len(df[df['y'] == 1]):
        # –¥–æ–±–∞–≤–∏—Ç—å –æ–±—ä–µ–∫—Ç–æ–≤ –ø–µ—Ä–≤–æ–≥–æ –∫–ª–∞—Å—Å–∞,
        df_new = pd.concat([df[df['y'] == 0],
                            df[df['y'] == 1].sample(n=len(df[df['y'] == 0]), replace=True, random_state=random_state_global)])
    else:
        # –∏–Ω–∞—á–µ –¥–æ–±–∞–≤–∏—Ç—å –æ–±—ä–µ–∫—Ç–æ–≤ –Ω—É–ª–µ–≤–æ–≥–æ –∫–ª–∞—Å—Å–∞.
        df_new = pd.concat([df[df['y'] == 1],
                            df[df['y'] == 0].sample(n=len(df[df['y'] == 1]), replace=True, random_state=random_state_global)])
    
    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤.
    df_new = df_new.sample(frac=1, random_state=random_state_global)
    
    return df_new.drop(columns=['y']), df_new['y']
````

### –¢–µ—Ö–Ω–∏–∫–∞ "downsampling"

#### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è downsampling.
def downsampling(X, y):
    df = X.copy()
    df['y'] = y.copy()
    
    # –ï—Å–ª–∏ –Ω—É–ª–µ–≤–æ–π –∫–ª–∞—Å—Å –ø—Ä–µ–æ–±–ª–∞–¥–∞–µ—Ç,
    if len(df[df['y'] == 0]) > len(df[df['y'] == 1]):
        # —É–º–µ–Ω—å—à–∏—Ç—å –µ–≥–æ –¥–æ —Ä–∞–∑–º–µ—Ä–æ–≤ –ø–µ—Ä–≤–æ–≥–æ –∫–ª–∞—Å—Å–∞,
        df_new = pd.concat([df[df['y'] == 0].sample(n=len(df[df['y'] == 1]), random_state=random_state_global),
                            df[df['y'] == 1]])
    else:
        # –∏–Ω–∞—á–µ —É–º–µ–Ω—å—à–∏—Ç—å –ø–µ—Ä–≤—ã–π –∫–ª–∞—Å—Å –¥–æ —Ä–∞–∑–º–µ—Ä–æ–≤ –Ω—É–ª–µ–≤–æ–≥–æ.
        df_new = pd.concat([df[df['y'] == 1].sample(n=len(df[df['y'] == 0]), random_state=random_state_global),
                            df[df['y'] == 0]])
    
    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤.
    df_new = df_new.sample(frac=1, random_state=random_state_global)
    
    return df_new.drop(columns=['y']), df_new['y']
````
